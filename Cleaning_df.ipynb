```python
import pandas as pd
import pandas_profiling
import numpy as np
```

```python
data = pd.read_csv('risk_factors_cervical_cancer.csv')
```

```python
pd.set_option('display.max_columns', 500)
```

```python
data = data.replace('?', 0)
```

```python
data = data.apply(pd.to_numeric)
```

ADDING COLUMNS TO DF

```python
data['Smoke > 10 years'] = np.where(data['Smokes (years)'] > 10, 1, 0)
```

```python
data['Been Pregnant'] = np.where(data['Num of pregnancies'] > 0, 1, 0)
data['Known First Sexual Intercourse Age'] = np.where(data['First sexual intercourse'] != 0, 1, 0)
```

```python
data['> 1 STD'] = np.where(data['STDs (number)'] > 1, 1, 0)
```

```python
data['>10 Years Sexual Activity'] = np.where(data['Age'] - data['First sexual intercourse'] >= 10, 1, 0)
```

```python
data['IUD >= 5 Years'] = np.where(data['IUD (years)'] >= 5, 1, 0)
```

```python
data['>= 3 Pregnancies'] = np.where(data['Num of pregnancies'] >= 3, 1, 0)
```

```python
#drop time since columns due to lack of values
data = data.drop(['STDs: Time since first diagnosis', 'STDs: Time since last diagnosis'], axis = 1)
```
